```
import numpy as np
t=np.arange(0.,5.,0.2)
plt.plot(t, t, 'or--', t, t**4, 'b*', t, t**3, 'gs')                     #gs=green square , b^=blue triangle , b*=blue star, r--=red line strips, o=circles
plt.show()
```

```
data = {'a': np.arange(50),
        'c': np.random.randint(0, 20, 50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

plt.scatter('a', 'c', c='d', s='d', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```

```
names=['group_a','group_b','group_c']
values=[1,10,100]
plt.figure(figsize=(9, 3))
plt.subplot(131)
plt.bar(names, values)
plt.subplot(132)
plt.scatter(names, values)
plt.subplot(133)
plt.plot(names, values)
plt.title('Bar graph')
plt.suptitle('Categorical Plotting')
plt.title('line graph')
plt.show()
```

```
mu, sigma = 100, 10
x = mu + sigma * np.random.randn(10000)
n, bins, patches = plt.hist(x, 50, density=1, facecolor='k', alpha=0.75)                #histrogram of the  data

plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60, .025, r'$\mu=100,\ \sigma=15$')
plt.axis([20, 460, 0, 0.05])
plt.grid(True)
plt.show()
```

```
import matplotlib.pyplot as plt
plt.style.use('classic')
%matplotlib inline
import numpy as np
x = np.linspace(0, 10, 1000)
fig, ax = plt.subplots()
ax.plot(x, np.sin(x), '-k', label='Sine')
ax.plot(x, np.cos(x), '--g', label='Cos')
ax.axis('equal')
leg = ax.legend();

  
```

```
ax.legend(loc='lower right', frameon=False)
fig                                                           #here by using lengend lable ---sine and --cos(lable)will diaplay on lower right side
```
